```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib

# Load csv
file_path = 'all fees.csv'
data = pd.read_csv(file_path)

# Clean column names
data.columns = data.columns.str.strip()

# Summary stats
summary_stats = data.describe(include='all')
print("Summary Statistics:")
print(summary_stats)

# Data types and missing values
print("\nData Info:")
data.info()

```

```python
# Distribution of fees (after cleaning non-numeric values)
data['Fee'] = data['Fee'].replace('[^\d.]', '', regex=True)
data['Fee'] = pd.to_numeric(data['Fee'], errors='coerce')
fee_distribution = data['Fee'].describe()
print("\n Fee Distribution:")
print(fee_distribution)

# Count of courses by university
courses_by_university = data['University'].value_counts()
print("\n Courses by University:")
print(courses_by_university)

# Distribution of UG and PG courses
ug_pg_distribution = data['UG/PG'].value_counts()
print("\n UG and PG Course Distribution:")
print(ug_pg_distribution)

# Visualization of fee distribution
plt.figure(figsize=(10, 6))
data['Fee'].hist(bins=50)
plt.title('Fee Distribution')
plt.xlabel('Fee')
plt.ylabel('Frequency')
plt.grid(False)
plt.show()

# Visualization of number of courses by university
plt.figure(figsize=(12, 8))
courses_by_university.plot(kind='bar')
plt.title('Number of Courses by University')
plt.xlabel('University')
plt.ylabel('Number of Courses')
plt.xticks(rotation=90)
plt.show()

# Visualization of UG and PG distribution
plt.figure(figsize=(8, 6))
ug_pg_distribution.plot(kind='bar')
plt.title('Distribution of UG and PG Courses')
plt.xlabel('Course Type')
plt.ylabel('Number of Courses')
plt.show()

```

```python

from nltk.corpus import stopwords
import matplotlib.pyplot as plt


# Calculate Q1 (25th percent) and Q3 (75th percentile)
Q1 = data['Fee'].quantile(0.25)
Q3 = data['Fee'].quantile(0.75)

# Calc iqr
IQR = Q3 - Q1

# Define the lower and upper bounds for detecting outliers
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Outlier detection in fees
Q1 = data['Fee'].quantile(0.25)
Q3 = data['Fee'].quantile(0.75)
IQR = Q3 - Q1
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Identify outlier fees
outliers = data[(data['Fee'] < lower_bound) | (data['Fee'] > upper_bound)]
print("\nOutlier Fees:")
print(outliers[['University', 'Course Name', 'Fee', 'UG/PG']])

# Save outliers to a CSV file
outliers.to_csv('outliers.csv', index=False)
print("\nOutliers saved to 'outliers.csv'")


```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt


file_path = 'all fees.csv'
data = pd.read_csv(file_path)

# Clean column names
data.columns = data.columns.str.strip()

# Convert fees to numeric after cleaning non-numeric values
data['Fee'] = data['Fee'].replace('[^\d.]', '', regex=True)
data['Fee'] = pd.to_numeric(data['Fee'], errors='coerce')

# Calculate average fees by course level and uni
average_fees = data.groupby(['University', 'UG/PG'])['Fee'].mean().reset_index()

# Display the average fees
print("Average Fees by University and Course Level:")
print(average_fees)
average_fees.to_csv('average fees.csv', index=False)

# Plot box and whisker plot of fees by course level and university
plt.figure(figsize=(15, 10))
sns.boxplot(x='University', y='Fee', hue='UG/PG', data=data)
plt.xticks(rotation=90)
plt.title('Box and Whisker Plot of Fees by Course Level and University')
plt.xlabel('University')
plt.ylabel('Fee')
plt.show()

# Violin plot of fees by course level and university
plt.figure(figsize=(15, 10))
sns.violinplot(x='University', y='Fee', hue='UG/PG', data=data)
plt.xticks(rotation=90)
plt.title('Violin Plot of Fees by Course Level and University')
plt.xlabel('University')
plt.ylabel('Fee')
plt.show()

```
